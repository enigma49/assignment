```python
import requests
import bs4
import lxml
import re
import io
```

```python
data = requests.get("http://www.espncricinfo.com/ci/content/player/index.html")
data = bs4.BeautifulSoup(data.text,'lxml')
mainurl = "http://www.espncricinfo.com/ci/content/player/caps.html?country="
total = data.select(".ciPlayersHomeCtryList")
for i in total:
    teams = i.find_all("a")

count = len(teams)
for j in range(count):
    url = teams[j].get("href")
   #print(url)
    url = re.split("=",url)
    url1 = mainurl+url[1]+";class=2"
    teamname = teams[j].text
    teamdata = requests.get(url1)
    teamdata = bs4.BeautifulSoup(teamdata.text,"lxml")
    playerdata = teamdata.select(".ciPlayerbycapstable")
    with io.open(teamname+".csv","w",encoding="utf8") as f1:
        f1.write("TEAMNAME,NAME,DEBUTDATE \n")
        f1.close()
   #print(len(playerdata))
    
    for k in playerdata:
        playerstat = k.find_all("li",class_="sep")
        for n in playerstat:
            number = n.find_all("li",class_ ="ciPlayerbycapstable")
            if(len(number)<1):
                number = n.find_all("li",class_="ciPlayerserielno")
              # print(number[0].text)
                name = n.find_all("li",class_="ciPlayername")
                print(name[0].text)
                debut = n.find_all("li",class_="ciPlayerplayed")
                print(debut[0].text)
                dataline = teamname +","+ name[0].text +","+ debut[0].text
                with io.open(teamname+".csv","a",encoding="utf8") as f1:
                        f1.write(dataline + "\n")
                        f1.close()
```
